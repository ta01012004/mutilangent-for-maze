```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import colors

class DualAgentMaze:
    def __init__(self, maze_size=(10, 10), start_nodes=None, goal_nodes=None):
        self.maze_size = maze_size
        self.width = int(self.maze_size[0])
        self.height = int(self.maze_size[1])
        self.maze_map = np.zeros((self.height, self.width), dtype='int')

        # Đặt các điểm bắt đầu cho hai tác nhân
        if start_nodes is None:
            self.start_nodes = [(0, 0), (0, self.width - 1)]  # Hai tác nhân ở hai góc trên
        else:
            self.start_nodes = start_nodes

        # Đặt các điểm mục tiêu (phần thưởng) ở vị trí khác nhau
        if goal_nodes is None:
            self.goal_nodes = [(self.height - 1, 0), (self.height - 1, self.width - 1)]  # Hai phần thưởng ở hai góc dưới
        else:
            self.goal_nodes = goal_nodes

        # Đặt tường cố định để tạo ra mê cung
        self.add_fixed_walls()

    def add_fixed_walls(self):
        """ Thêm các tường cố định vào mê cung để tạo ra độ khó cao hơn và loại bỏ tường cụ thể """
        # Tường được thêm vào để tạo ra các lối đi hẹp và hành lang
        walls = [
            # Tạo các bức tường để tạo thành lối đi
            (1, 1), (1, 2), (1, 4), (1, 5), (1, 7),
            (2, 3), (2, 7), (2, 8),
            (3, 1), (3, 5), (3, 6), (3, 8),
            (4, 0), (4, 3), (4, 7), (4, 9),
            (5, 2), (5, 5), (5, 6),
            (6, 1), (6, 4), (6, 8),
            (7, 3), (7, 7),
            (8, 2), (8, 6), (8, 9)
        ]

        # Loại bỏ tường cụ thể mà bạn yêu cầu
        walls_to_remove = [(6, 4)]  # Bức tường tại vị trí này sẽ được loại bỏ
        walls = [wall for wall in walls if wall not in walls_to_remove]

        for (i, j) in walls:
            self.maze_map[i, j] = 1  # Đánh dấu tường là 1

    def plot_maze(self, agents_positions=None, show=True):
        # Tạo bản sao của `maze_map` để không thay đổi bản gốc
        temp_maze_map = np.copy(self.maze_map)

        # Đặt các tác nhân tại vị trí hiện tại của chúng
        if agents_positions:
            for idx, pos in enumerate(agents_positions):
                i, j = pos
                temp_maze_map[i, j] = 2 + idx  # Đặt màu xanh dương cho Agent A và màu xanh lá cho Agent B

        # Đặt điểm mục tiêu
        for (i, j) in self.goal_nodes:
            temp_maze_map[i, j] = 4  # Đánh dấu mục tiêu chung là 4 (màu vàng)

        # Tạo màu sắc cho từng giá trị trong bản đồ mê cung
        color_map = {
            0: 'white',     # Đường đi
            1: 'black',     # Tường
            2: 'blue',      # Vị trí hiện tại của tác nhân A
            3: 'green',     # Vị trí hiện tại của tác nhân B
            4: 'yellow',    # Phần thưởng chung cho cả hai tác nhân
        }

        # Chuẩn bị bản đồ màu cho `matplotlib`
        img = np.zeros((temp_maze_map.shape[0], temp_maze_map.shape[1], 3))
        for (i, j), value in np.ndenumerate(temp_maze_map):
            img[i, j] = colors.to_rgb(color_map[value])

        # Hiển thị mê cung với kích thước vừa phải và tường sát với tác nhân
        plt.figure(figsize=(8, 8))
        plt.imshow(img, origin='lower')

        plt.xticks([])
        plt.yticks([])
        plt.grid(visible=False)  # Tắt lưới để chỉ tập trung vào các thành phần của mê cung

        if show:
            plt.show()
```

# Chạy ngẫu nhiên

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import colors
import random

# Lớp Agent đại diện cho mỗi tác nhân trong mê cung
class Agent:
    def __init__(self, name, start_position, goal_position):
        self.name = name
        self.start_position = start_position
        self.position = start_position
        self.goal = goal_position
        self.total_steps = 0  # Đếm số bước đã đi

    def reset(self):
        """Khởi tạo lại trạng thái của tác nhân"""
        self.position = self.start_position
        self.total_steps = 0

    def step(self, new_position):
        """Cập nhật vị trí mới của tác nhân"""
        self.position = new_position
        self.total_steps += 1

# Lớp DualAgentMaze đã được giữ nguyên như trước đây.

# Lớp môi trường mê cung cho hai tác nhân
class MazeEnv:
    def __init__(self, maze):
        self.maze = maze
        # Khởi tạo hai tác nhân với vị trí bắt đầu và mục tiêu
        self.agent_A = Agent("Agent A", maze.start_nodes[0], maze.goal_nodes[0])
        self.agent_B = Agent("Agent B", maze.start_nodes[1], maze.goal_nodes[1])
        self.agents = [self.agent_A, self.agent_B]
        self.goals_reached = [False, False]  # Theo dõi đích nào đã bị chiếm

    def reset(self):
        """Khởi tạo lại trạng thái ban đầu của môi trường và cả hai tác nhân"""
        self.agent_A.reset()
        self.agent_B.reset()
        self.goals_reached = [False, False]
        return [self.agent_A.position, self.agent_B.position]

    def step(self, agent_idx, action):
        """Thực hiện hành động của một tác nhân và nhận lại trạng thái mới, phần thưởng, và trạng thái kết thúc"""
        agent = self.agents[agent_idx]
        possible_actions = self.get_possible_actions(agent)
        reward = -0.1  # Phần thưởng tiêu cực cho mỗi bước di chuyển để khuyến khích tìm đường nhanh nhất

        if action in possible_actions:
            agent.step(action)  # Cập nhật vị trí mới của tác nhân
        else:
            reward = -1  # Nếu hành động không hợp lệ, phần thưởng tiêu cực lớn hơn

        done = False
        # Kiểm tra nếu tác nhân đến đích của nó và đích chưa bị chiếm
        if agent.position == agent.goal and not self.goals_reached[agent_idx]:
            reward = 1  # Phần thưởng khi đến đích
            done = True  # Kết thúc nếu tác nhân đến đích
            self.goals_reached[agent_idx] = True  # Đánh dấu đích đã bị chiếm

            # Khi đích của một tác nhân đã bị chiếm, đích đó sẽ không còn là mục tiêu cho tác nhân khác
            for other_agent_idx, other_agent in enumerate(self.agents):
                if other_agent_idx != agent_idx and other_agent.goal == agent.goal:
                    self.goals_reached[other_agent_idx] = True

        return agent.position, reward, done

    def get_possible_actions(self, agent):
        """Lấy các hành động có thể từ vị trí hiện tại của tác nhân"""
        current_x, current_y = agent.position
        possible_actions = []

        # Các bước di chuyển có thể (lên, xuống, trái, phải)
        directions = [(-1, 0), (1, 0), (0, -1), (0, 1)]
        for dx, dy in directions:
            nx, ny = current_x + dx, current_y + dy
            if 0 <= nx < self.maze.height and 0 <= ny < self.maze.width:
                if self.maze.maze_map[nx, ny] != 1:  # Không phải tường
                    possible_actions.append((nx, ny))

        return possible_actions

    def render(self):
        """Hiển thị mê cung và vị trí của hai tác nhân"""
        agents_positions = [agent.position for agent in self.agents]
        self.maze.plot_maze(agents_positions=agents_positions)

# Chạy một tập (episode) với hai tác nhân trong môi trường
def run_dual_agent_episode(env):
    """Chạy một tập (episode) với hai tác nhân trong môi trường"""
    states = env.reset()
    done = [False, False]
    total_reward = [0, 0]

    while not all(done):
        env.render()  # Hiển thị mê cung và tác nhân ở mỗi bước

        for agent_idx in range(2):
            if not done[agent_idx]:
                # Chọn một hành động ngẫu nhiên cho tác nhân
                possible_actions = env.get_possible_actions(env.agents[agent_idx])
                action = random.choice(possible_actions)

                # Thực hiện hành động của tác nhân
                next_state, reward, done[agent_idx] = env.step(agent_idx, action)
                total_reward[agent_idx] += reward

                print(f"{env.agents[agent_idx].name} moved to {next_state}, Reward: {reward}, Total steps: {env.agents[agent_idx].total_steps}")

    print(f"Episode finished.")
    for agent_idx in range(2):
        print(f"{env.agents[agent_idx].name} total reward: {total_reward[agent_idx]}, Total steps: {env.agents[agent_idx].total_steps}")

# Khởi tạo và chạy tập
if __name__ == "__main__":
    # Khởi tạo đối tượng `DualAgentMaze`
    dual_agent_maze = DualAgentMaze(maze_size=(10, 10))
    env = MazeEnv(dual_agent_maze)
    run_dual_agent_episode(env)
```

# Q-leaning

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import colors
import random
from collections import defaultdict
from tabulate import tabulate  # Thêm thư viện tabulate để in bảng đẹp mắt
import time  # Thêm thư viện time để theo dõi thời gian huấn luyện

# Lớp QLearningAgent sử dụng Q-learning và thuật toán hợp tác
class QLearningAgent:
    def __init__(self, name, start_position, goal_position, actions, agent_number, shared_q_table,
                 gamma=0.9, learning_rate=0.1, exploration_rate=0.1, exploration_decay=0.995, delta=0.1):
        self.name = name
        self.start_position = start_position
        self.position = start_position
        self.goal = goal_position
        self.actions = actions  # Các hành động có thể: lên, xuống, trái, phải
        self.gamma = gamma  # Hệ số chiết khấu γ
        self.learning_rate = learning_rate
        self.exploration_rate = exploration_rate  # Epsilon cho chính sách epsilon-greedy
        self.exploration_decay = exploration_decay  # Tỷ lệ giảm epsilon
        self.delta = delta  # Số δ (>0)
        self.q_table = shared_q_table  # Sử dụng bảng Q chung
        self.total_steps = 0
        self.agent_number = agent_number  # Số lượng tác nhân

        # Khởi tạo MaxStep và sử dụng trong defaultdict
        self.MaxStep = 1000
        self.tix = defaultdict(lambda: self.MaxStep)  # Lưu trữ tix_i cho mỗi mục tiêu x
        self.Tix = defaultdict(lambda: self.MaxStep)  # Lưu trữ Tix cho mỗi mục tiêu x
        self.internal_reward = 0  # Phần thưởng nội bộ

        # Khởi tạo tix và Tix cho mục tiêu riêng của tác nhân
        self.tix[self.goal] = self.MaxStep
        self.Tix[self.goal] = self.MaxStep

    def reset(self):
        """Khởi tạo lại trạng thái của tác nhân."""
        self.position = self.start_position
        self.total_steps = 0
        self.internal_reward = 0
        # Không reset tix và Tix để giữ lại thông tin tốt nhất

    def choose_action(self, possible_actions):
        """Chọn một hành động dựa trên chính sách epsilon-greedy."""
        if random.uniform(0, 1) < self.exploration_rate:
            # Khám phá: chọn hành động ngẫu nhiên
            return random.choice(possible_actions)
        else:
            # Khai thác: chọn hành động tốt nhất từ Q-table
            q_values = [self.q_table[(self.position, action)] for action in possible_actions]
            max_q = max(q_values)
            # Trong trường hợp có nhiều hành động có cùng giá trị Q lớn nhất, chọn ngẫu nhiên một trong số chúng
            max_actions = [action for action, q in zip(possible_actions, q_values) if q == max_q]
            return random.choice(max_actions)

    def update_q_value(self, state, action, reward, next_state, next_possible_actions):
        """Cập nhật giá trị Q dựa trên chuyển tiếp quan sát được và phần thưởng nội bộ."""
        # Tính toán phần thưởng tổng hợp
        total_reward = reward + self.internal_reward
        current_q = self.q_table[(state, action)]
        if next_possible_actions:
            next_q_values = [self.q_table[(next_state, next_action)] for next_action in next_possible_actions]
            max_next_q = max(next_q_values)
        else:
            max_next_q = 0
        # Quy tắc cập nhật Q-learning
        new_q = current_q + self.learning_rate * (total_reward + self.gamma * max_next_q - current_q)
        self.q_table[(state, action)] = new_q

    def decay_exploration(self):
        """Giảm tỷ lệ khám phá."""
        self.exploration_rate *= self.exploration_decay

    def estimate_internal_reward(self):
        """Tính toán phần thưởng nội bộ dựa trên công thức."""
        tAL = self.tix[self.goal]  # Thời gian tốt nhất đã biết
        tAS = self.Tix[self.goal]  # Thời gian thực tế trong tập hiện tại
        gamma = self.gamma
        delta = self.delta
        if tAL - tAS > 0:
            ir = (gamma ** (tAL - tAS)) * self.internal_reward + delta
        else:
            ir = self.internal_reward
        self.internal_reward = ir

    def share_tix(self, agents):
        """Chia sẻ tix_i với các tác nhân khác."""
        for agent in agents:
            if agent != self:
                if self.goal in agent.tix:
                    agent.tix[self.goal] = min(agent.tix[self.goal], self.tix[self.goal])

# Lớp DualAgentMaze giữ nguyên như trước đây
class DualAgentMaze:
    def __init__(self, maze_size=(10, 10), start_nodes=None, goal_nodes=None):
        self.maze_size = maze_size
        self.width = int(self.maze_size[0])
        self.height = int(self.maze_size[1])
        self.maze_map = np.zeros((self.height, self.width), dtype='int')

        # Đặt các điểm bắt đầu cho hai tác nhân
        if start_nodes is None:
            self.start_nodes = [(0, 0), (0, self.width - 1)]  # Hai tác nhân ở hai góc trên
        else:
            self.start_nodes = start_nodes

        # Đặt các điểm mục tiêu ở vị trí khác nhau
        if goal_nodes is None:
            self.goal_nodes = [(self.height - 1, 0), (self.height - 1, self.width - 1)]  # Hai mục tiêu ở hai góc dưới
        else:
            self.goal_nodes = goal_nodes

        # Đặt tường cố định để tạo ra mê cung
        self.add_fixed_walls()

    def add_fixed_walls(self):
        """Thêm các tường cố định vào mê cung để tạo ra độ khó cao hơn"""
        # Tường được thêm vào để tạo ra các lối đi hẹp và hành lang
        walls = [
            # Tạo các bức tường để tạo thành lối đi
            (1, 1), (1, 2), (1, 4), (1, 5), (1, 7),
            (2, 3), (2, 7), (2, 8),
            (3, 1), (3, 5), (3, 6), (3, 8),
            (4, 0), (4, 3), (4, 7), (4, 9),
            (5, 2), (5, 5), (5, 6),
            (6, 1), (6, 4), (6, 8),
            (7, 3), (7, 7),
            (8, 2), (8, 6), (8, 9)
        ]

        # Loại bỏ tường cụ thể nếu cần
        walls_to_remove = [(6, 4)]  # Bức tường tại vị trí này sẽ được loại bỏ
        walls = [wall for wall in walls if wall not in walls_to_remove]

        for (i, j) in walls:
            self.maze_map[i, j] = 1  # Đánh dấu tường là 1

    def plot_maze(self, agents_positions=None, show=True):
        # Tạo bản sao của `maze_map` để không thay đổi bản gốc
        temp_maze_map = np.copy(self.maze_map)

        # Đặt các tác nhân tại vị trí hiện tại của chúng
        if agents_positions:
            for idx, pos in enumerate(agents_positions):
                i, j = pos
                temp_maze_map[i, j] = 2 + idx  # Đặt màu cho mỗi tác nhân

        # Đặt điểm mục tiêu
        for (i, j) in self.goal_nodes:
            temp_maze_map[i, j] = 4  # Đánh dấu mục tiêu là 4 (màu vàng)

        # Tạo màu sắc cho từng giá trị trong bản đồ mê cung
        color_map = {
            0: 'white',     # Đường đi
            1: 'black',     # Tường
            2: 'blue',      # Agent A
            3: 'green',     # Agent B
            4: 'yellow',    # Mục tiêu
        }

        # Chuẩn bị bản đồ màu cho `matplotlib`
        img = np.zeros((temp_maze_map.shape[0], temp_maze_map.shape[1], 3))
        for (i, j), value in np.ndenumerate(temp_maze_map):
            img[i, j] = colors.to_rgb(color_map[value])

        # Hiển thị mê cung
        plt.figure(figsize=(6, 6))
        plt.imshow(img, origin='lower')

        plt.xticks([])
        plt.yticks([])
        plt.grid(visible=False)

        if show:
            plt.show()

# Lớp môi trường
class MazeEnv:
    def __init__(self, maze):
        self.maze = maze
        # Các hành động: lên, xuống, trái, phải
        self.actions = [(-1, 0), (1, 0), (0, -1), (0, 1)]
        # Tạo một bảng Q chung
        shared_q_table = defaultdict(float)
        # Khởi tạo hai tác nhân Q-learning sử dụng bảng Q chung
        self.agent_A = QLearningAgent(
            "Agent A",
            maze.start_nodes[0],
            maze.goal_nodes[0],
            self.actions,
            agent_number=2,
            shared_q_table=shared_q_table
        )
        self.agent_B = QLearningAgent(
            "Agent B",
            maze.start_nodes[1],
            maze.goal_nodes[1],
            self.actions,
            agent_number=2,
            shared_q_table=shared_q_table
        )
        self.agents = [self.agent_A, self.agent_B]
        self.goals_reached = [False, False]

    def reset(self):
        for agent in self.agents:
            agent.reset()
        self.goals_reached = [False, False]
        return [agent.position for agent in self.agents]

    def step(self, agent_idx, action):
        agent = self.agents[agent_idx]
        possible_actions = self.get_possible_actions(agent)
        reward = -0.1  # Phạt cho mỗi bước đi

        if action in possible_actions:
            new_position = (agent.position[0] + action[0], agent.position[1] + action[1])
            agent.position = new_position
            agent.total_steps += 1
        else:
            reward = -1  # Phạt nặng cho hành động không hợp lệ

        done = False
        if agent.position == agent.goal and not self.goals_reached[agent_idx]:
            reward = 1  # Phần thưởng khi đến đích
            done = True
            self.goals_reached[agent_idx] = True

            # Cập nhật Tix
            agent.Tix[agent.goal] = agent.total_steps

        return agent.position, reward, done

    def get_possible_actions(self, agent):
        x, y = agent.position
        possible_actions = []
        for dx, dy in self.actions:
            nx, ny = x + dx, y + dy
            if 0 <= nx < self.maze.height and 0 <= ny < self.maze.width:
                if self.maze.maze_map[nx, ny] != 1:  # Không phải tường
                    possible_actions.append((dx, dy))
        return possible_actions

    def render(self):
        """Hiển thị mê cung và vị trí của hai tác nhân"""
        agents_positions = [agent.position for agent in self.agents]
        self.maze.plot_maze(agents_positions=agents_positions)

# Hàm huấn luyện các tác nhân với thuật toán hợp tác và thu thập phần thưởng, thời gian
def train_agents(env, episodes=1000, max_steps_per_episode=1000, moving_average_window=100):
    # Khởi tạo tix_i cho mỗi tác nhân
    for agent in env.agents:
        agent.tix[agent.goal] = agent.MaxStep

    # Khởi tạo danh sách lưu trữ phần thưởng và thời gian
    rewards_history = {agent.name: [] for agent in env.agents}
    moving_averages = {agent.name: [] for agent in env.agents}
    steps_history = []  # Lưu trữ số bước mỗi tập
    time_history = []   # Lưu trữ thời gian mỗi tập

    for episode in range(1, episodes + 1):
        env.reset()
        done = [False, False]
        steps = 0
        episode_rewards = {agent.name: 0 for agent in env.agents}

        start_time = time.time()  # Bắt đầu đếm thời gian cho tập này

        for agent in env.agents:
            agent.decay_exploration()  # Giảm tỷ lệ khám phá

        while not all(done) and steps < max_steps_per_episode:
            steps += 1
            for idx, agent in enumerate(env.agents):
                if not done[idx]:
                    state = agent.position
                    possible_actions = env.get_possible_actions(agent)
                    if not possible_actions:
                        continue  # Không có hành động khả thi
                    action = agent.choose_action(possible_actions)
                    next_state, reward, done[idx] = env.step(idx, action)
                    next_possible_actions = env.get_possible_actions(agent)
                    agent.update_q_value(state, action, reward, next_state, next_possible_actions)

                    # Ghi nhận phần thưởng
                    episode_rewards[agent.name] += reward

                    # Nếu tác nhân đã đến đích
                    if done[idx]:
                        # Cập nhật Tix
                        agent.Tix[agent.goal] = steps

            # Sau mỗi bước, các tác nhân chia sẻ tix_i
            for agent in env.agents:
                goal = agent.goal
                Tix = agent.Tix[goal]
                if Tix < agent.tix[goal]:
                    agent.tix[goal] = Tix
                    agent.share_tix(env.agents)  # Chia sẻ tix_i với các tác nhân khác

        end_time = time.time()  # Kết thúc đếm thời gian cho tập này
        elapsed_time = end_time - start_time  # Thời gian đã trôi qua cho tập này

        # Sau mỗi tập, các tác nhân ước tính phần thưởng nội bộ
        for agent in env.agents:
            agent.estimate_internal_reward()

        # Ghi nhận phần thưởng và số bước của mỗi tác nhân trong tập này
        for agent in env.agents:
            rewards_history[agent.name].append(episode_rewards[agent.name])

            # Tính trung bình di động nếu đủ số tập
            if episode >= moving_average_window:
                avg_reward = np.mean(rewards_history[agent.name][-moving_average_window:])
                moving_averages[agent.name].append(avg_reward)
            else:
                moving_averages[agent.name].append(np.mean(rewards_history[agent.name]))

        # Ghi nhận số bước và thời gian cho tập này
        steps_history.append(steps)
        time_history.append(elapsed_time)

        # In thông tin sau mỗi khoảng số tập đã định (ví dụ: sau mỗi 100 tập)
        if episode % moving_average_window == 0 or episode == 1:
            print(f"Tập {episode}/{episodes} - Phần thưởng trung bình (trên {moving_average_window} tập gần nhất):")
            table = []
            for agent in env.agents:
                avg_reward = moving_averages[agent.name][-1]
                table.append([agent.name, round(avg_reward, 4)])
            print(tabulate(table, headers=["Agent", f"Average Reward (trên {moving_average_window} tập gần nhất)"], tablefmt="pretty"))
            print(f"Số bước trung bình mỗi tập: {np.mean(steps_history[-moving_average_window:]):.2f}")
            print(f"Thời gian trung bình mỗi tập: {np.mean(time_history[-moving_average_window:]):.4f} giây/tập")
            print("-" * 80)

    print("Huấn luyện hoàn tất.")
    return rewards_history, moving_averages, steps_history, time_history

# Hàm để in bảng trung bình phần thưởng và thời gian sau huấn luyện
def print_average_reward_table(moving_averages, moving_average_window=100):
    table = []
    for agent_name, rewards in moving_averages.items():
        final_avg = rewards[-1]
        table.append([agent_name, f"{final_avg:.4f}"])
    print("\nBảng Trung Bình Phần Thưởng Cuối Cùng:")
    print(tabulate(table, headers=["Agent", f"Average Reward (trên {moving_average_window} tập gần nhất)"], tablefmt="pretty"))

def print_average_time_table(time_history, moving_average_window=100):
    # Tính trung bình thời gian trên các tập gần nhất
    if len(time_history) >= moving_average_window:
        avg_time = np.mean(time_history[-moving_average_window:])
    else:
        avg_time = np.mean(time_history)
    print("\nBảng Trung Bình Thời Gian Huấn Luyện:")
    print(tabulate([["Average Training Time", f"{avg_time:.4f} giây/tập"]], headers=["Metric", "Value"], tablefmt="pretty"))

# Hàm để vẽ biểu đồ phần thưởng trung bình di động
def plot_moving_averages(moving_averages, moving_average_window=100):
    plt.figure(figsize=(12, 6))
    for agent_name, averages in moving_averages.items():
        plt.plot(averages, label=agent_name)
    plt.xlabel('Tập Huấn Luyện')
    plt.ylabel('Phần Thưởng Trung Bình')
    plt.title(f'Phần Thưởng Trung Bình Di Động (Trên {moving_average_window} Tập Gần Nhất)')
    plt.legend()
    plt.grid(True)
    plt.show()

# Hàm để vẽ biểu đồ thời gian huấn luyện trung bình
def plot_training_time(time_history, moving_average_window=100):
    moving_avg_time = []
    for i in range(len(time_history)):
        if i+1 >= moving_average_window:
            avg_time = np.mean(time_history[i+1 - moving_average_window:i+1])
        else:
            avg_time = np.mean(time_history[:i+1])
        moving_avg_time.append(avg_time)

    plt.figure(figsize=(12, 6))
    plt.plot(moving_avg_time, label='Average Training Time per Episode')
    plt.xlabel('Tập Huấn Luyện')
    plt.ylabel('Thời Gian Huấn Luyện Trung Bình (giây)')
    plt.title(f'Thời Gian Huấn Luyện Trung Bình Di Động (Trên {moving_average_window} Tập Gần Nhất)')
    plt.legend()
    plt.grid(True)
    plt.show()

# Chạy các tác nhân đã được huấn luyện
def run_trained_agents(env):
    env.reset()
    done = [False, False]
    total_reward = {agent.name: 0 for agent in env.agents}

    while not all(done):
        env.render()
        for idx, agent in enumerate(env.agents):
            if not done[idx]:
                state = agent.position
                possible_actions = env.get_possible_actions(agent)
                if not possible_actions:
                    continue  # Không có hành động khả thi
                # Chọn hành động tốt nhất dựa trên Q-values đã học
                q_values = [agent.q_table[(state, action)] for action in possible_actions]
                max_q = max(q_values)
                max_actions = [action for action, q in zip(possible_actions, q_values) if q == max_q]
                action = random.choice(max_actions)
                next_state, reward, done[idx] = env.step(idx, action)
                total_reward[agent.name] += reward
                print(f"{agent.name} moved to {next_state}, Reward: {reward}, Total steps: {agent.total_steps}")

    env.render()
    for agent in env.agents:
        print(f"{agent.name} tổng phần thưởng: {total_reward[agent.name]}, Tổng số bước: {agent.total_steps}")

# Hàm chính
if __name__ == "__main__":
    # Khởi tạo đối tượng `DualAgentMaze`
    maze = DualAgentMaze(maze_size=(10, 10))
    env = MazeEnv(maze)

    # Huấn luyện các tác nhân và thu thập lịch sử phần thưởng và thời gian
    rewards_history, moving_averages, steps_history, time_history = train_agents(
        env, episodes=1000, moving_average_window=100
    )

    # In bảng trung bình phần thưởng và thời gian huấn luyện cuối cùng
    print_average_reward_table(moving_averages, moving_average_window=100)
    print_average_time_table(time_history, moving_average_window=100)

    # Vẽ biểu đồ phần thưởng trung bình di động
    plot_moving_averages(moving_averages, moving_average_window=100)

    # Vẽ biểu đồ thời gian huấn luyện trung bình di động
    plot_training_time(time_history, moving_average_window=100)

    # Chạy các tác nhân đã được huấn luyện
    run_trained_agents(env)


```

```python
# Lớp QLearningAgent sử dụng Q-learning và thuật toán hợp tác
class QLearningAgent:
    def __init__(self, name, start_position, goal_position, actions, agent_number, shared_q_table,
                 gamma=0.9, learning_rate=0.1, exploration_rate=0.9, exploration_decay=0.995, delta=0.1):
        self.name = name
        self.start_position = start_position
        self.position = start_position
        self.goal = goal_position
        self.actions = actions  # Các hành động có thể: lên, xuống, trái, phải
        self.gamma = gamma  # Hệ số chiết khấu γ
        self.learning_rate = learning_rate
        self.exploration_rate = exploration_rate  # Epsilon cho chính sách epsilon-greedy
        self.exploration_decay = exploration_decay  # Tỷ lệ giảm epsilon
        self.delta = delta  # Số δ (>0)
        self.q_table = shared_q_table  # Sử dụng bảng Q chung
        self.total_steps = 0
        self.agent_number = agent_number  # Số lượng tác nhân

        # Khởi tạo MaxStep và sử dụng trong defaultdict
        self.MaxStep = 1000
        self.tix = defaultdict(lambda: self.MaxStep)  # Lưu trữ tix_i cho mỗi mục tiêu x
        self.Tix = defaultdict(lambda: self.MaxStep)  # Lưu trữ Tix cho mỗi mục tiêu x
        self.internal_reward = 0  # Phần thưởng nội bộ

        # Khởi tạo tix và Tix cho mục tiêu riêng của tác nhân
        self.tix[self.goal] = self.MaxStep
        self.Tix[self.goal] = self.MaxStep

    def reset(self):
        """Khởi tạo lại trạng thái của tác nhân."""
        self.position = self.start_position
        self.total_steps = 0
        self.internal_reward = 0
        # Không reset tix và Tix để giữ lại thông tin tốt nhất

    def choose_action(self, possible_actions):
        """Chọn một hành động dựa trên chính sách epsilon-greedy."""
        if random.uniform(0, 1) < self.exploration_rate:
            # Khám phá: chọn hành động ngẫu nhiên
            return random.choice(possible_actions)
        else:
            # Khai thác: chọn hành động tốt nhất từ Q-table
            q_values = [self.q_table[(self.position, action)] for action in possible_actions]
            max_q = max(q_values)
            # Trong trường hợp có nhiều hành động có cùng giá trị Q lớn nhất, chọn ngẫu nhiên một trong số chúng
            max_actions = [action for action, q in zip(possible_actions, q_values) if q == max_q]
            return random.choice(max_actions)

    def update_q_value(self, state, action, reward, next_state, next_possible_actions):
        """Cập nhật giá trị Q dựa trên chuyển tiếp quan sát được và phần thưởng nội bộ."""
        # Tính toán phần thưởng tổng hợp
        total_reward = reward + self.internal_reward
        current_q = self.q_table[(state, action)]
        if next_possible_actions:
            next_q_values = [self.q_table[(next_state, next_action)] for next_action in next_possible_actions]
            max_next_q = max(next_q_values)
        else:
            max_next_q = 0
        # Quy tắc cập nhật Q-learning
        new_q = current_q + self.learning_rate * (total_reward + self.gamma * max_next_q - current_q)
        self.q_table[(state, action)] = new_q

    def decay_exploration(self):
        """Giảm tỷ lệ khám phá."""
        self.exploration_rate *= self.exploration_decay

    def estimate_internal_reward(self):
        """Tính toán phần thưởng nội bộ dựa trên công thức."""
        tAL = self.tix[self.goal]  # Thời gian tốt nhất đã biết
        tAS = self.Tix[self.goal]  # Thời gian thực tế trong tập hiện tại
        gamma = self.gamma
        delta = self.delta
        if tAL - tAS > 0:
            ir = (gamma ** (tAL - tAS)) * self.internal_reward + delta
        else:
            ir = self.internal_reward
        self.internal_reward = ir

    def share_tix(self, agents):
        """Chia sẻ tix_i với các tác nhân khác."""
        for agent in agents:
            if agent != self:
                if self.goal in agent.tix:
                    agent.tix[self.goal] = min(agent.tix[self.goal], self.tix[self.goal])

```

```python
# Hàm huấn luyện các tác nhân với thuật toán hợp tác và thu thập tổng phần thưởng, thời gian
def train_agents(env, episodes=1000, max_steps_per_episode=1000):
    # Khởi tạo tix_i cho mỗi tác nhân
    for agent in env.agents:
        agent.tix[agent.goal] = agent.MaxStep

    # Khởi tạo danh sách lưu trữ tổng phần thưởng và thời gian
    rewards_history = {agent.name: 0 for agent in env.agents}  # Lưu tổng phần thưởng cho mỗi tác nhân
    steps_history = []  # Lưu trữ số bước mỗi tập
    total_time = 0  # Tổng thời gian huấn luyện

    for episode in range(1, episodes + 1):
        env.reset()
        done = [False, False]
        steps = 0
        episode_rewards = {agent.name: 0 for agent in env.agents}

        start_time = time.time()  # Bắt đầu đếm thời gian cho tập này

        for agent in env.agents:
            agent.decay_exploration()  # Giảm tỷ lệ khám phá

        while not all(done) and steps < max_steps_per_episode:
            steps += 1
            for idx, agent in enumerate(env.agents):
                if not done[idx]:
                    state = agent.position
                    possible_actions = env.get_possible_actions(agent)
                    if not possible_actions:
                        continue  # Không có hành động khả thi
                    action = agent.choose_action(possible_actions)
                    next_state, reward, done[idx] = env.step(idx, action)
                    next_possible_actions = env.get_possible_actions(agent)
                    agent.update_q_value(state, action, reward, next_state, next_possible_actions)

                    # Ghi nhận phần thưởng cho tập này
                    episode_rewards[agent.name] += reward

                    # Nếu tác nhân đã đến đích
                    if done[idx]:
                        # Cập nhật Tix
                        agent.Tix[agent.goal] = steps

            # Sau mỗi bước, các tác nhân chia sẻ tix_i
            for agent in env.agents:
                goal = agent.goal
                Tix = agent.Tix[goal]
                if Tix < agent.tix[goal]:
                    agent.tix[goal] = Tix
                    agent.share_tix(env.agents)  # Chia sẻ tix_i với các tác nhân khác

        end_time = time.time()  # Kết thúc đếm thời gian cho tập này
        elapsed_time = end_time - start_time  # Thời gian đã trôi qua cho tập này
        total_time += elapsed_time  # Cộng dồn tổng thời gian huấn luyện

        # Sau mỗi tập, các tác nhân ước tính phần thưởng nội bộ
        for agent in env.agents:
            agent.estimate_internal_reward()

        # Ghi nhận tổng phần thưởng của mỗi tác nhân sau mỗi tập
        for agent in env.agents:
            rewards_history[agent.name] += episode_rewards[agent.name]  # Cộng dồn tổng phần thưởng

        # Ghi nhận số bước của tập này
        steps_history.append(steps)

        # In thông tin sau mỗi tập
        if episode > 0:
            print(f"Tập {episode}/{episodes} - Tổng phần thưởng:")
            table = []
            for agent in env.agents:
                total_reward = rewards_history[agent.name]  # Sử dụng rewards_history thay vì rewards_
                table.append([agent.name, round(total_reward, 4)])
            print(tabulate(table, headers=["Agent", "Total Reward"], tablefmt="pretty"))
            print(f"Số bước trung bình mỗi tập: {np.mean(steps_history):.2f}")
            print(f"Tổng thời gian huấn luyện: {total_time:.4f} giây")
            print("-" * 80)

    print("Huấn luyện hoàn tất.")
    return rewards_history, total_time, steps_history

```

```python
# Hàm để in bảng trung bình phần thưởng và thời gian sau huấn luyện
def print_average_reward_table(moving_averages, moving_average_window=100):
    table = []
    for agent_name, rewards in moving_averages.items():
        final_avg = rewards[-1]
        table.append([agent_name, f"{final_avg:.4f}"])
    print("\nBảng Trung Bình Phần Thưởng Cuối Cùng:")
    print(tabulate(table, headers=["Agent", f"Average Reward (trên {moving_average_window} tập gần nhất)"], tablefmt="pretty"))

def print_average_time_table(time_history, moving_average_window=100):
    # Tính trung bình thời gian trên các tập gần nhất
    if len(time_history) >= moving_average_window:
        avg_time = np.mean(time_history[-moving_average_window:])
    else:
        avg_time = np.mean(time_history)
    print("\nBảng Trung Bình Thời Gian Huấn Luyện:")
    print(tabulate([["Average Training Time", f"{avg_time:.4f} giây/tập"]], headers=["Metric", "Value"], tablefmt="pretty"))

# Hàm để in ra tổng phần thưởng của bảng Q chung
def print_q_table_summary(shared_q_table):
    total_q_sum = sum(shared_q_table.values())
    num_entries = len(shared_q_table)
    average_q = total_q_sum / num_entries if num_entries > 0 else 0
    max_q = max(shared_q_table.values()) if num_entries > 0 else 0
    min_q = min(shared_q_table.values()) if num_entries > 0 else 0

    summary_table = [
        ["Total Q-value", f"{total_q_sum:.4f}"],
        ["Number of State-Action Pairs", num_entries],
        ["Average Q-value", f"{average_q:.4f}"],
        ["Maximum Q-value", f"{max_q:.4f}"],
        ["Minimum Q-value", f"{min_q:.4f}"]
    ]

    print("\nBảng Tổng Quan Bảng Q Chung:")
    print(tabulate(summary_table, headers=["Metric", "Value"], tablefmt="pretty"))

# Hàm để vẽ biểu đồ phần thưởng trung bình di động
def plot_moving_averages(moving_averages, moving_average_window=100):
    plt.figure(figsize=(12, 6))
    for agent_name, averages in moving_averages.items():
        plt.plot(averages, label=agent_name)
    plt.xlabel('Tập Huấn Luyện')
    plt.ylabel('Phần Thưởng Trung Bình')
    plt.title(f'Phần Thưởng Trung Bình Di Động (Trên {moving_average_window} Tập Gần Nhất)')
    plt.legend()
    plt.grid(True)
    plt.show()

# Hàm để vẽ biểu đồ thời gian huấn luyện trung bình di động
def plot_training_time(time_history, moving_average_window=100):
    moving_avg_time = []
    for i in range(len(time_history)):
        if i+1 >= moving_average_window:
            avg_time = np.mean(time_history[i+1 - moving_average_window:i+1])
        else:
            avg_time = np.mean(time_history[:i+1])
        moving_avg_time.append(avg_time)

    plt.figure(figsize=(12, 6))
    plt.plot(moving_avg_time, label='Average Training Time per Episode')
    plt.xlabel('Tập Huấn Luyện')
    plt.ylabel('Thời Gian Huấn Luyện Trung Bình (giây)')
    plt.title(f'Thời Gian Huấn Luyện Trung Bình Di Động (Trên {moving_average_window} Tập Gần Nhất)')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
# Chạy các tác nhân đã được huấn luyện
def run_trained_agents(env):
    env.reset()
    done = [False, False]
    total_reward = {agent.name: 0 for agent in env.agents}

    while not all(done):
        env.render()
        for idx, agent in enumerate(env.agents):
            if not done[idx]:
                state = agent.position
                possible_actions = env.get_possible_actions(agent)
                if not possible_actions:
                    continue  # Không có hành động khả thi
                # Chọn hành động tốt nhất dựa trên Q-values đã học
                q_values = [agent.q_table[(state, action)] for action in possible_actions]
                max_q = max(q_values)
                max_actions = [action for action, q in zip(possible_actions, q_values) if q == max_q]
                action = random.choice(max_actions)
                next_state, reward, done[idx] = env.step(idx, action)
                total_reward[agent.name] += reward
                print(f"{agent.name} moved to {next_state}, Reward: {reward}, Total steps: {agent.total_steps}")

    env.render()
    for agent in env.agents:
        print(f"{agent.name} tổng phần thưởng: {total_reward[agent.name]}, Tổng số bước: {agent.total_steps}")
```

```python
import os
import pickle

# Hàm để lưu lại bảng Q của các tác nhân sau khi huấn luyện
def save_model(env, filename='trained_model.pkl'):
    # Kiểm tra xem thư mục /mnt/data/ có tồn tại không, nếu không thì tạo nó
    os.makedirs('/mnt/data/', exist_ok=True)

    # Lưu bảng Q của từng tác nhân
    model_data = {agent.name: agent.q_table for agent in env.agents}
    with open('/mnt/data/' + filename, 'wb') as f:
        pickle.dump(model_data, f)

    print(f"Model saved to /mnt/data/{filename}")

# Sau khi huấn luyện xong, gọi hàm lưu mô hình
save_model(env, 'trained_model.pkl')

```
